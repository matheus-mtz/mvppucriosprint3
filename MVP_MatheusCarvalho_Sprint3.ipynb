## ***2. Objetivos***
---

Claro, aqui estão alguns objetivos mais específicos voltados para a construção do Data Lake e análises de reviews por produtos e data das vendas:

1. **Coleta de Dados Eficiente:** Garantir que o Data Lake seja capaz de coletar e armazenar de maneira eficiente todos os reviews de clientes relacionados a produtos da Amazon, incluindo informações sobre a data das vendas e os produtos adquiridos.

2. **Integração de Dados de Múltiplas Fontes:** Integrar dados de avaliações de clientes com informações adicionais, como dados de vendas e dados de produtos, para fornecer uma visão holística e enriquecida das análises.

3. **Armazenamento Escalável:** Construir o Data Lake com a capacidade de lidar com grandes volumes de dados à medida que mais avaliações são adicionadas ao longo do tempo, garantindo escalabilidade.

4. **Agrupamento de Reviews por Produto:** Organizar as avaliações em grupos por produtos específicos, permitindo análises detalhadas de feedback para cada produto.

5. **Segmentação de Produtos e Análise Comparativa:** Segmentar produtos com base em critérios como categoria, preço e marca, e comparar seu desempenho com base na data das vendas.

Estes objetivos direcionam a construção do Data Lake e as análises específicas que podem ser realizadas para obter insights valiosos sobre como as avaliações de clientes, produtos e datas de venda se relacionam e impactam o desempenho da Amazon.


## ***3. Detalhamento da base de dados***
---

Este é um banco de dados de reviews de produtos da Amazon que contém informações detalhadas sobre os produtos e as avaliações dos clientes. Aqui está uma descrição dos campos disponíveis no banco de dados:

- id (object): Um identificador único para cada entrada no banco de dados.

- dateAdded (object): A data em que a entrada foi adicionada ao banco de dados.

- dateUpdated (object): A data em que a entrada foi atualizada pela última vez no banco de dados.

- name (object): O nome do produto.

- asins (object): Números ASIN (Amazon Standard Identification Number) associados ao produto.

- brand (object): A marca do produto.

- categories (object): Categorias às quais o produto pertence.

- primaryCategories (object): A categoria principal do produto.

- imageURLs (object): URLs das imagens do produto.

- keys (object): Chaves relacionadas ao produto.

- manufacturer (object): O fabricante do produto.

- manufacturerNumber (object): O número de fabricação do produto atribuído pelo fabricante.

- reviews.date (object): A data em que a revisão foi escrita pelo cliente.

- reviews.dateAdded (object): A data em que a revisão foi adicionada ao banco de dados.

- reviews.dateSeen (object): A data em que a revisão foi vista pela primeira vez.

- reviews.doRecommend (bool): Um valor booleano que indica se o revisor recomenda o produto ou não.

- reviews.id (float64): Um identificador único para cada revisão.

- reviews.numHelpful (int64): O número de pessoas que acharam a revisão útil.

- reviews.rating (int64): A classificação atribuída pelo revisor ao produto.

- reviews.sourceURLs (object): URLs das fontes das revisões.

- reviews.text (object): O texto da revisão escrita pelo cliente.

- reviews.title (object): O título da revisão, se fornecido pelo cliente.

- reviews.username (object): O nome de usuário do revisor.

- sourceURLs (object): URLs das páginas onde as reviews foram colhidas.

Este banco de dados é valioso para análises de produtos, monitoramento de avaliações de clientes e pesquisa de mercado na plataforma da Amazon. Ele fornece informações detalhadas sobre os produtos, suas avaliações e as opiniões dos clientes, permitindo uma análise abrangente das tendências e desempenho dos produtos na plataforma.

## ***7. Análise***
---

Nesta seção, vamos implantar uma solução usando Google Colab para instanciar uma sessão Spark, importar os dados e realizar a contagem.

```python
!pip install pyspark --quiet
```

```python
from pyspark import *
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master("local[*]").getOrCreate()

conf = SparkConf().set('spark.iu.port','4050').setAppName("reviews").setMaster("local[*]")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
rdf = spark.read
    .option('header','true')  #Indica que a primeira linha contém cabeçalhos de coluna
    .option('inferSchema',true)  #Inferir automaticamente os tipos de dados das colunas
    .csv(url)
```

```python
rdf.write.saveAsTable("reviews")
```

```python
spark.sql("SELECT name, COUNT(*) FROM reviews GROUP BY name").show()
```
